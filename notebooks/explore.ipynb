# Making the Dataset Tidy

```python
def clean_cafes(df):
    """Clean the cafes dataset"""
    # remove duplicates
    new_df = df.drop_duplicates(subset=["Census year", "Trading name", "Property ID"])

    # make Seating Type a column variable by setting all cols except for 'Number of Seats' as 
    # indexes, then unstacking using 'Seating Type'
    index_cols = list(df.columns)
    index_cols.remove('Number of seats')
    new_df = new_df.set_index(index_cols).unstack("Seating type", fill_value=0).reset_index(col_level=1)
    new_df.columns = new_df.columns.droplevel(0)
    return new_df


cleaned_cafes_df = clean_cafes(cafes_df)
cleaned_cafes_df.head()
```

# Plotting with Pandas (Matplotlib)

Let's use Pandas to plot the number of records for each year. 

Note that Pandas' plotting API uses Matplotlib by default.

```python
years_df = cafes_df["Census year"].value_counts().sort_index().rename_axis("year").reset_index(name="records")
years_df.head()
```

```python
%matplotlib inline
years_df.plot.bar(x="year", title="Venue counts by years", figsize=(15,8))
```

# Plotting with Matplotlib directly

```python
import matplotlib.pyplot as plt

# Set params of the current plot (that has yet to be drawn)
fig = plt.figure(figsize=(15, 8))
plt.title("Venue counts by years")
plt.xlabel("year")
plt.legend(["records"])

# make the plot
ax = plt.bar(x=years_df["year"], height=years_df["records"]);
```

# Seaborn

Seaborn is a wrapper around Matplotlib wich offers the following:

* a more intuitive API than Matplotlib
* offers a range of out-of-the-box plots for doing statistical analysis
* makes prettier plots

```python
import seaborn as sns

sns.set(style="whitegrid")

sns.set(rc={'figure.figsize':(15, 10)})

sns.barplot(x="year", y="records", data=years_df).set_title("Venue counts by years")
```
