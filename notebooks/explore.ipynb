# Making the Dataset Tidy

```python
def clean_cafes(df):
    """Clean the cafes dataset"""
    # remove duplicates
    new_df = df.drop_duplicates(subset=["Census year", "Trading name", "Property ID"])

    # make Seating Type a column variable by setting all cols except for 'Number of Seats' as 
    # indexes, then unstacking using 'Seating Type'
    index_cols = list(df.columns)
    index_cols.remove('Number of seats')
    new_df = new_df.set_index(index_cols).unstack("Seating type", fill_value=0).reset_index(col_level=1)
    new_df.columns = new_df.columns.droplevel(0)
    return new_df


cleaned_cafes_df = clean_cafes(cafes_df)
cleaned_cafes_df.head()
```

# Plotting with Pandas (Matplotlib)

Let's use Pandas to plot the number of records for each year. 

Note that Pandas' plotting API uses Matplotlib by default.

```python
years_df = cleaned_cafes_df["Census year"].value_counts().sort_index().rename_axis("year").reset_index(name="records")
years_df
```

```python
%matplotlib inline
years_df.plot.bar(x="year", title="Venue counts by years", figsize=(15,8))
```

```python
def make_facet_plot(df, year=2017, cutoff=100, seat_type="Seats - Indoor"):
    # filter to year   
    cafes_2017_df = df[df["Census year"] == year]

    # prune the long tail
    cafes_2017_df = cafes_2017_df[cafes_2017_df[seat_type] < cutoff]
    
    # Initialize the FacetGrid object
    pal = sns.cubehelix_palette(10, rot=-.25, light=.7)
    cat = sns.FacetGrid(cafes_2017_df, row="CLUE small area", hue="CLUE small area", aspect=15, height=1, palette=pal)

    # Draw the densities in a few steps
    cat.map(sns.kdeplot, seat_type, clip_on=False, shade=True, alpha=1, lw=1.5, bw=.2)
    cat.map(sns.kdeplot, seat_type, clip_on=False, color="w", lw=2, bw=.2)
    cat.map(plt.axhline, y=0, lw=2, clip_on=False)

    # Define and use a simple function to label the plot in axes coordinates
    def label(x, color, label):
        ax = plt.gca()
        ax.text(1, .2, label, fontweight="bold", color=color,
                ha="right", va="top", transform=ax.transAxes)

    cat.map(label, "Seats - Indoor")

    # Set the subplots to overlap
    cat.fig.subplots_adjust(hspace=0)

    # Remove axes details that don't play well with overlap
    cat.set_titles("")
    cat.set(yticks=[])
    cat.despine(bottom=True, left=True)

    
make_facet_plot(cleaned_cafes_df);
```
